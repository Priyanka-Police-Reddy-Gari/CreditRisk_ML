```python
final_data=pd.DataFrame()
data["S_2"]=pd.to_datetime(data["S_2"])
#data["Year-Month"] = data["S_2"].dt.to_period('M')
#data1=data.groupby('customer_ID').apply(lambda x: x.sample(n=1)).reset_index(drop = True)
final_data=pd.concat([final_data,pd.merge(labels,data, how= 'inner', on='customer_ID')])
final_data
```

One hot encoding

```python
#One-Hot
final_data = pd.get_dummies(final_data, columns = ['D_63', 'D_64'])
print(final_data)
```

```python
final_data[['B_30','B_38','D_114','D_116','D_117','D_120','D_126','D_66','D_68']]
# final_data['D_68'].unique()
# final_data.shape
```

```python
for col in['B_30','B_38','D_114','D_116','D_117','D_120','D_126','D_66','D_68']:
    col_dummies=pd.get_dummies(final_data[col],prefix=col)
    final_data=pd.concat([final_data,col_dummies],axis=1)
    # print(final_data.shape)
    final_data.drop(col,axis=1,inplace=True)
print(final_data.shape)
```

Feature Engineering

```python
final_data.describe().transpose()
```

```python
features = final_data.drop([ 'S_2' ], axis=1).columns.to_list()
# print(final_data['customer_ID'])
# print(features)
cat_col = [i for i in features if i.startswith('B_30')or i.startswith('B_38') or i.startswith('D_114') or i.startswith('D_116')or i.startswith('D_117') or i.startswith('D_120') or i.startswith('D_126') or i.startswith('D_63')or i.startswith('D_64')or i.startswith('D_66') or i.startswith('D_68')]
print(cat_col)
ind_col = ['customer_ID', 'target']
print(ind_col)
# cat_col = ['B_30', 'B_38', 'D_114', 'D_116', 'D_117', 'D_120', 'D_126', 'D_63', 'D_64', 'D_66', 'D_68']
num_col = [col for col in features if col not in cat_col and col not in ind_col]
print('NUM COL',num_col)
```

```python
df_num_agg = final_data.groupby('customer_ID')[num_col].agg(['mean', 'std', 'min', 'max', 'last'])
df_num_agg.columns = ['_'.join(x) for x in df_num_agg.columns]
print(df_num_agg)
print(df_num_agg.shape)
```
